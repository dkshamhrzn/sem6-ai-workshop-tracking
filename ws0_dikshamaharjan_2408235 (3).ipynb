```python
def length_conversion(fromto, value):
    """
        Converts length from meters to feet or vice versa.
        parameters: fromto:string, value:float
        returns: None
    """
    try:
        if fromto == "m":
          print(value * 3.28084, "ft")
        elif fromto == "ft":
         print(value / 3.28084, "m")
    except:
        print("Invalid unit. Please enter 'm' or 'ft'")

def volume_conversion(fromto, value):
    """
        Converts volume from liters to gallons or vice versa.
        parameters: fromto:string, value:float
        returns: None
    """
    if fromto == "l":
        print(value * 0.264172, "gal")
    elif fromto == "gal":
        print(value / 0.264172, "l")
    else:
        print("Invalid unit. Please enter 'l' or 'gal'")

def weight_conversion(fromto, value):
    """
        Converts weight from kilograms to pounds or vice versa.
        parameters: fromto:string, value:float
        returns: None
    """
    if fromto == "kg":
        print(value * 2.20462, "lbs")
    elif fromto == "lbs":
        print(value / 2.20462, "kg")
    else:
        print("Invalid unit. Please enter 'kg' or 'lbs'")


con = input("Enter type of conversion (length, volume, weight): ")

if con == "length":
    fromto = input("Enter unit to convert from (m or ft): ")
    value = float(input("Enter value: "))
    length_conversion(fromto, value)

elif con == "volume":
    fromto = input("Enter unit to convert from (l or gal): ")
    value = float(input("Enter value: "))
    volume_conversion(fromto, value)

elif con == "weight":
    fromto = input("Enter unit to convert from (kg or lbs): ")
    value = float(input("Enter value: "))
    weight_conversion(fromto, value)

else:
    print("Invalid conversion type.")
```

```python
def calc_sum(numbers):
    """
    Calculates the sum of a list of numbers.
    Parameters: numbers (list of float)
    Returns: float: The sum of all numbers in the list.

    """
    return sum(numbers)


def calc_average(numbers):
    """
    Calculates the average (mean) of a list of numbers.
    Parameters: numbers (list of float)
    Returns: float: The arithmetic mean of all numbers in the list.
    """
    return sum(numbers) / len(numbers)


def calc_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.
    Parameters: numbers (list of float)
    Returns:float: The largest number in the list.
    """
    return max(numbers)


def calc_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.
    Parameters:numbers (list of float): A non-empty list of numeric values.
    Returns: float: The smallest number in the list.
    """
    return min(numbers)


def get_numbers():
    """
    Prompts the user to enter a space-separated list of numbers
    and converts them into a list of floats.
    Parameters: None
    Returns: list of float: A list of numbers entered by the user.
    error catch:
        ValueError: If the input is empty or contains non-numeric values.
    """
    raw = input("Enter numbers separated by spaces: ")
    if not raw.strip():
        raise ValueError("No numbers entered. List cannot be empty.")
    return [float(x) for x in raw.split()]


def main():
    """
    Main function that displays a menu, takes user input,
    and calls the appropriate mathematical operation function.
    Parameters: None
    Returns: None — results are printed directly to the console.
    Error: ValueError: Caught internally if input is invalid or list is empty.
    """
    print("Math Operations:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")
    print("\n")

    try:
        choice = input("Choose an operation (1/2/3/4): ").strip()

        if choice not in ("1", "2", "3", "4"):
            print("Invalid choice. Please enter 1, 2, 3, or 4.")
            return

        numbers = get_numbers()

        if choice == "1":
            print(f"Sum: {calc_sum(numbers)}")
        elif choice == "2":
            print(f"Average: {calc_average(numbers)}")
        elif choice == "3":
            print(f"Maximum: {calc_maximum(numbers)}")
        elif choice == "4":
            print(f"Minimum: {calc_minimum(numbers)}")

    except ValueError as e:
        print(f"Error: {e}")


main()
```

```python
# 4.2 Exercise on List Manipulation


# Question 1: Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from the first."""
    return lst[::2]

lst1 = [1, 2, 3, 4, 5, 6]
print("Q1 - Every other element:", extract_every_other(lst1))


# Question 2: Slice a Sublist
def get_sublist(lst, start, end):
    """Returns a sublist from start to end index (inclusive)."""
    return lst[start:end + 1]

lst2 = [1, 2, 3, 4, 5, 6]
print("Q2 - Sublist (start=2, end=4):", get_sublist(lst2, 2, 4))


# Question 3: Reverse a List Using Slicing
def reverse_list(lst):
    """Returns a reversed list using slicing."""
    return lst[::-1]

lst3 = [1, 2, 3, 4, 5]
print("Q3 - Reversed list:", reverse_list(lst3))


# Question 4: Remove the First and Last Elements
def remove_first_last(lst):
    """Returns the list without the first and last elements."""
    return lst[1:-1]

lst4 = [1, 2, 3, 4, 5]
print("Q4 - Remove first & last:", remove_first_last(lst4))


# Question 5: Get the First n Elements
def get_first_n(lst, n):
    """Returns the first n elements of the list."""
    return lst[:n]

lst5 = [1, 2, 3, 4, 5]
print("Q5 - First 3 elements:", get_first_n(lst5, 3))


# Question 6: Extract Elements from the End
def get_last_n(lst, n):
    """Returns the last n elements of the list."""
    return lst[-n:]

lst6 = [1, 2, 3, 4, 5]
print("Q6 - Last 2 elements:", get_last_n(lst6, 2))


# Question 7: Extract Elements in Reverse Order
def reverse_skip(lst):
    """Returns every second element in reverse, starting from second-to-last."""
    return lst[-2::-2]

lst7 = [1, 2, 3, 4, 5, 6]
print("Q7 - Reverse skip:", reverse_skip(lst7))
```

Exercises on nested list

```python

# Question 1: Flatten a Nested List
def flatten(lst):
    """Flattens a one-level nested list into a single list."""
    return [elem for sublist in lst for elem in sublist]

lst1 = [[1, 2], [3, 4], [5]]
print("Q1 - Flattened list:", flatten(lst1))


# Question 2: Accessing Nested List Elements
def access_nested_element(lst, indices):
    """Returns the element at the given indices in a nested list."""
    for i in indices:
        lst = lst[i]
    return lst

lst2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("Q2 - Accessed element:", access_nested_element(lst2, [1, 2]))


# Question 3: Sum of All Elements in a Nested List
def sum_nested(lst):
    """Returns the sum of all elements in a nested list of any depth."""
    total = 0
    for elem in lst:
        if isinstance(elem, list):
            total += sum_nested(elem)
        else:
            total += elem
    return total

lst3 = [[1, 2], [3, [4, 5]], 6]
print("Q3 - Sum of nested list:", sum_nested(lst3))


# Question 4: Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from a nested list."""
    return [[x for x in sublist if x != elem] for sublist in lst]

lst4 = [[1, 2], [3, 2], [4, 5]]
print("Q4 - After removing 2:", remove_element(lst4, 2))


# Question 5: Find the Maximum Element in a Nested List
def find_max(lst):
    """Returns the maximum element in a nested list of any depth."""
    max_val = None
    for elem in lst:
        if isinstance(elem, list):
            candidate = find_max(elem)
        else:
            candidate = elem
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val

lst5 = [[1, 2], [3, [4, 5]], 6]
print("Q5 - Maximum element:", find_max(lst5))


# Question 6: Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """Counts how many times elem appears in a nested list."""
    count = 0
    for sublist in lst:
        count += sublist.count(elem)
    return count

lst6 = [[1, 2], [2, 3], [2, 4]]
print("Q6 - Count of 2:", count_occurrences(lst6, 2))


# Question 7: Flatten a List of Lists of Lists
def deep_flatten(lst):
    """Flattens a deeply nested list of any depth into a single list."""
    result = []
    for elem in lst:
        if isinstance(elem, list):
            result.extend(deep_flatten(elem))
        else:
            result.append(elem)
    return result

lst7 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print("Q7 - Deep flattened list:", deep_flatten(lst7))


# Question 8: Nested List Average
def average_nested(lst):
    """Returns the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

lst8 = [[1, 2], [3, 4], [5, 6]]
print("Q8 - Average of nested list:", average_nested(lst8))
```
